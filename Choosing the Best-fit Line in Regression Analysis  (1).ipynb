# Understanding the shapes of different types of function graphs is necessary in order to choose the best method for drawing the best-fit line that intersects the points during regression analysis.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = x**2
plt.figure(figsize=(4, 2))
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadratic Function')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = -x**2
plt.figure(figsize=(4, 2))  
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)  
y = np.sqrt(x)
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Square Root Function')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)  
y = -(np.sqrt(x))
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Square Root Function')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = x**3
plt.figure(figsize=(4, 2))  
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Cubic Function')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 100)
y = -x**3
plt.figure(figsize=(4, 2))  
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Cubic  Function')
plt.grid(True)
plt.show()
```
